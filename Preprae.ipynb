```python
import pandas as pd
import numpy as np
import os
```

```python
data_folder = './data'
data_format = 'csv'
print(f'All files in folder: {os.listdir(data_folder)}')
files = []

for file in os.listdir(data_folder):
    if file.endswith(data_format):
        files.append(file)
print(f'Filtred files: {files}')
```

```python
file_path =os.path.join(data_folder, files[0])
df = pd.read_csv(file_path)
```

```python
df.tail()
```

```python
df = df[df['game_type'] != 'doubles']
df = df[df['gender'] != 'women']
df=df.dropna()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.to_csv('./exports/scraped_cleaned.csv', index=False)
```

```python
df_old = pd.read_csv('./exports/combined_data_cleared_enhanced.csv')
```

```python
print(df.columns)
print(df_old.columns)
```

```python
df = df.rename(columns={'player1': 'Winner', 'player2': 'Loser', 'player1AtpRank': 'WRank', 'player2AtpRank': 'LRank', 'player1AtpPoints': 'WPts', 'player2AtpPoints': 'LPts', 'odd1': 'B365W', 'odd2': 'B365L' })
```

```python
result_df = pd.concat([df_old, df])
result_df = result_df.drop(columns=['game_type','competition','gender','date'])
```
